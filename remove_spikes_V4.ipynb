```python
d.process_diode()

averageDiodeTime = []
for i in range(len(d.diode_pulses())):
    averageDiodeTime.append((d.diode_pulses()[i][1] + d.diode_pulses()[i][0]) / 2)

diode_r_MJD = []
for i in range(len(d[551].mean(axis=1))):
    diode_r_MJD.append(0)
    for j in range(len(averageDiodeTime)):
        if i == (averageDiodeTime[j] - 0.5):
            #print("i = %f" %i)
            #print("averageDiodeTime[%f] = %f" % (j, averageDiodeTime[j]))
            diode_r_Index = (int)((i-(averageDiodeTime[0]-0.5))/160)
            #print("diode_r_Index = %f" % diode_r_Index)
            diode_r_MJD[i] = d.diode_r[551].mean(axis=1)[diode_r_Index]
x = []
y = []
for i in range(len(diode_r_MJD)):
    if diode_r_MJD[i] != 0:
        x.append(i)
        y.append(diode_r_MJD[i])    
plt.plot(x, y)
```
