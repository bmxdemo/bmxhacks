```python
da=bmxdata.BMXFile("/astro/u/anze/bmxdata/bmxdata/raw/1906/190614_1951_pulse__D2.data")
```

```python
plt.figure(figsize=(8,8))
istart=800
iend=920
cut=da.data['chan3_0'][istart:iend,0:1560].T
norm = simple_norm(cut, 'sqrt')
plt.imshow(cut,interpolation='nearest',aspect='auto',
               origin='lower',extent=(0,(iend-istart)*da.deltaT,da.freq[0][0],da.freq[0][1560]),norm=LogNorm())
cbar = plt.colorbar()
cbar.set_label('spectrometer units', rotation=90)
plt.xlabel('time [s]',fontsize=14)
plt.ylabel('frequency [MHz]',fontsize=14)
plt.savefig ('droneswitch.png',dpi=300)
```
